Artificial Narrow Intelligence (ANI), also called weak AI, refers to systems designed for specific tasks.

```python
# Example code: Supervised Learning with Iris Dataset

from sklearn.datasets import load_iris
import pandas as pd

# Load iris dataset (supervised: features with labels)
iris = load_iris()
data = pd.DataFrame(iris.data, columns=iris.feature_names)
labels = iris.target
target_names = iris.target_names

# Display dataset info
print("Features (last 5 rows):\n", data.tail())
print("Labels (last 5):", labels[-5:])
print("Label names:", target_names)
```

## Linear Regression

```python
# Example code: Supervised Learning with Linear Regression on House Prices

import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Sample data: house sizes in square feet and prices in dollars
house_sizes = np.array([1000, 1200, 1400, 1600, 1800, 2000, 2200, 2400, 2600, 2800]).reshape(-1, 1)
house_prices = np.array([200000, 220000, 240000, 260000, 280000, 300000, 320000, 340000, 360000, 380000])

# Split data into training (80%) and testing (20%) sets
X_train, X_test, y_train, y_test = train_test_split(house_sizes, house_prices, test_size=0.2, random_state=42)

# Create and train the linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Predict prices for the test set
y_pred = model.predict(X_test)

# Calculate mean squared error to evaluate model accuracy
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse}")

# Visualize the results
plt.scatter(X_train, y_train, color='blue', label='Training data')
plt.scatter(X_test, y_test, color='green', label='Test data')
plt.plot(X_train, model.predict(X_train), color='red', label='Regression line')
plt.xlabel('House Size (sq ft)')
plt.ylabel('House Price ($)')
plt.title('Simple Linear Regression: House Price vs. Size')
plt.legend()
plt.show()

model.predict([[1500]])  # Predict price for a house of size 1500 sq ft
print("Predicted price for a 1500 sq ft house:", model.predict([[1500]])[0])

model.predict([[3000]])  # Predict price for a house of size 3000 sq ft
print("Predicted price for a 3000 sq ft house:", model.predict([[3000]])[0])

house_size = model.coef_[0]  # Coefficient for house size
print("Coefficient for house size:", house_size)

intercept = model.intercept_ # Intercept of the regression line
print("Intercept:", intercept)

# Price increase per square foot
price_increase_per_sqft = house_size
print("Price increase per square foot:", price_increase_per_sqft)

# Calculate the R-squared value to access model fit
r_squared = model.score(X_train, y_train)
print("R-squared value:", r_squared)
print("R-squared value score evaluation: ", "Good" if r_squared > 0.7 else "Poor")
print("MSE evaluation: ", "Good" if mse < 1000000 else "Poor")
```

```python
# Linear regression on advertising budget and sales

# Sample data: advertising budgets in thousands of dollars and sales in thousands of units
ad_budgets = np.array([50, 60, 70, 80, 90, 100]).reshape(-1, 1)
sales = np.array([150, 180, 210, 240, 270, 300])

# Split data into training (80%) and testing (20%) sets
X_train_ad, X_test_ad, y_train_ad, y_test_ad = train_test_split(ad_budgets, sales, test_size=0.2, random_state=42)

# Create and train the linear regression model
model_ad = LinearRegression()
model_ad.fit(X_train_ad, y_train_ad)

# Predict sales for the test set
y_pred_ad = model_ad.predict(X_test_ad)

# Calculate mean squared error for the test set
mse_ad = mean_squared_error(y_test_ad, y_pred_ad)
print(f"Mean Squared Error for Advertising Budget Model: {mse_ad}")

# Visualize the results
plt.scatter(X_train_ad, y_train_ad, color='blue', label='Training data')
plt.scatter(X_test_ad, y_test_ad, color='green', label='Test data')
plt.scatter(X_test_ad, y_pred_ad, color='orange', label='Test predictions', marker='x')
plt.plot(X_train_ad, model_ad.predict(X_train_ad), color='red', label='Regression line')
plt.xlabel('Advertising Budget (thousands of $)')
plt.ylabel('Sales (thousands of units)')
plt.title('Linear Regression: Sales vs. Advertising Budget')
plt.legend()
plt.show()

# Calculate the R-squared value for the advertising budget model
r_squared_ad = model_ad.score(X_train_ad, y_train_ad)
print("R-squared value for Advertising Budget Model:", r_squared_ad)
print("R-squared value score evaluation:", "Good" if r_squared_ad > 0.7 else "Poor")

# Predict sales for a new advertising budget
new_budget = 75  # in thousands of dollars
predicted_sales = model_ad.predict([[new_budget]])
print(f"Predicted sales for an advertising budget of ${new_budget}k: ${round(predicted_sales[0] * 1000)}k")
```
