```python
# The following code examples are simple practice regressions using different libraries and techniques.
# This script demonstrates a basic linear regression to predict project abandonment duration based on reasons.

# EXAMPLE 1: Linear Regression to predict average days developers abandon projects
import pandas as pd  # pandas: Data manipulation and analysis library
from sklearn.linear_model import LinearRegression  # sklearn: Machine learning library for regression
import matplotlib.pyplot as plt  # matplotlib: Visualization library

# Sample data: reasons and corresponding average days of abandonment
data = {
    'reason': ['Lack of time', 'Lack of interest', 'Technical difficulties', 'Poor planning'],
    'average_days': [10, 20, 30, 15]
}

# Create DataFrame: Organizes data into a tabular structure for analysis
df = pd.DataFrame(data)

# Convert categorical 'reason' to numerical codes: Assigns unique integers to each category for model input
df['reason_num'] = df['reason'].astype('category').cat.codes

# Define features (X) and target (y): X is the independent variable (reason_num), y is the dependent variable (average_days)
X = df[['reason_num']]  # Double brackets create a DataFrame for sklearn
y = df['average_days']

# Initialize and train the model: LinearRegression fits a line to predict y from X
model = LinearRegression()
model.fit(X, y)  # fit(): Trains the model on the data

# Predict: Uses the trained model to estimate average days for each reason
predicted_days = model.predict(X)
print("Predicted average days for each reason:")
for reason, days in zip(df['reason'], predicted_days):
    print(f"{reason}: {days:.2f} days")

# Visualize results: Creates a bar chart to display predictions
plt.bar(df['reason'], predicted_days, color='skyblue')  # bar(): Plots categorical data
plt.xlabel('Reason for Abandonment')  # Labels x-axis
plt.ylabel('Predicted Average Days')  # Labels y-axis
plt.title('Predicted Average Days for Project Abandonment Reasons')  # Sets chart title
plt.xticks(rotation=45)  # Rotates x-axis labels for readability
plt.tight_layout()  # Adjusts layout to prevent label cutoff
plt.show()  # Displays the plot
```
